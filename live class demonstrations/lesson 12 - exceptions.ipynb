```python
while True:
    try:
        x = input('give me a number: ')
        print(2 ** int(x))
        break
    except ValueError:
        print('that wasnt a number, try again')
```

```python
while True:
    try:
        x = input('give me a number: ')
        print(2 ** int(x))
        break
    except Exception as ex:
        print('oops, that didnt work', ex, type(ex))
```
