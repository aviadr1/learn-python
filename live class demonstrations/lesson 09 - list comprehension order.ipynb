```python
menu = {
    'steak' : 150,
    'pita' : 1.5,
    'hamburger' : 59,
    'chips' : 17,
    'shakshuka' : 32,
    'cake' : 25
}

order = ['steak', 'steak', 'chips', 'cake']
```

```python
total = 0
for item in order:
    price = menu[item]
    total += price
    
print(total)
```

```python
sum([menu[item] for item in order ])
```
